# Pre-processing PDF Files

## PyPDF2 Library

The library is case sensitive. Need to type "PyPDF2" exactly as show here.

```python
from PyPDF2 import PdfFileReader, PdfFileWriter # classes to read and write to pdf's
```

```python
file_path = r"C:\Users\Loube\Desktop\OCS Packet\SPC_Louis_Range_OPORD.pdf"
```

```python
file = open(file_path, "rb")
```

```python
reader = PyPDF2.PdfFileReader(file)
```

```python
page1 = reader.getPage(0)
```

```python
data = []
for page in range(reader.getNumPages()):
    page = reader.getPage(page)
    text = page.extractText().split("UNCLASSIFIED FOR TRAINING OPORD 21-09 (OPERATION GREEN MOTH)")
    print(text)
    print("\n")
    data.append(text)
```
