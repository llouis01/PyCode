```python
# import libraries
import PyPDF2 as pdf
import pandas as pd
```

```python
# import pdf file with python's open()
file = open(r"C:\Users\Loube\Desktop\OCS Packet\SPC_Louis_range_OPORD.pdf", "rb") # read binary
file
```

```python
pdf_reader = pdf.PdfFileReader(file) # pdf file reader
pdf_reader
```

```python
help(pdf_reader)
```

```python
pdf_reader.getIsEncrypted() # is doc encrypted?
```

```python
pdf_reader.getNumPages() # how many pages?
```

```python
page1 = pdf_reader.getPage(0)
page2 = pdf_reader.getPage(1)
page3 = pdf_reader.getPage(2)
page4 = pdf_reader.getPage(3)
```

```python
page2.extractText()
```

## Append Write or Merge PDF's

```python
pdf_writer = pdf.PdfFileWriter()
```

```python
pdf_writer.addPage(page1)
pdf_writer.addPage(page2)
```

```python
output = open("Pages.pdf", "wb")
pdf_writer.write(output)
output.close()
```
