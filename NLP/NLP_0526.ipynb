### Pre-processing text data

Cleaning up the text data is necessary to highlight attriutes we want the model to pick up on. Pre-processing of text data is consisted of three steps:

    1. Remove punctuation
    2. Tokenization
    3. Remove stopwords
    4. Lemmatize/Stem

```python
import pandas as pd
pd.set_option("display.max_colwidth", 100)

data = pd.read_csv("C://Users//Loube//Desktop//Ex_Files_NLP_Python_ML_EssT//Exercise Files//Ch01//01_03//Start//SMSSpamCollection.tsv", sep="\t", header=None)
data.columns = ['label', 'body_text']

data.head()
```

## Remove punctuation

```python
import string
string.punctuation
```

```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

data['body_text_clean'] = data['body_text'].apply(lambda x: remove_punct(x))

data.head()
```

## Tokenization

```python
import re

def tokenize(text):
    tokens = re.split("\W+", text)
    return tokens

data['body_text_tokenized'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))

data.head()
```

## Remove stopwords

```python
import nltk

stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text

data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

data.head()
```

# Stemming

### Stripping a word down to its base word.
-Marching -> March
-Flying -> Fly
Berries -> Berry

### Purpose is to reduce the corpus the model is exposed to.


```python
ps = nltk.PorterStemmer() # the Porter Stemmer, most popular stemmer
```

```python
dir(ps)
```

```python
print(ps.stem("growing"))
print(ps.stem("growes"))
print(ps.stem("grow"))
```

```python
print(ps.stem("run"))
print(ps.stem("running"))
print(ps.stem("runner"))
```

## Read in raw text

```python
stopwords = nltk.corpus.stopwords.words("english")
data = pd.read_csv(r"C:\Users\Loube\Desktop\Ex_Files_NLP_Python_ML_EssT\Exercise Files\Ch01\01_03\Start\SMSSpamCollection.tsv", sep="\t")
data.columns = ['label', 'body_text']
data.head()
```

```python
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text

data['body_text_nostop'] = data['body_text'].apply(lambda x: clean_text(x.lower()))

data.head()
```

# Stem text

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

data['body_text_stemmed'] = data['body_text_nostop'].apply(lambda x: stemming(x))

data.head()

# stemming explicitly correlates words to a similar meaning
```
